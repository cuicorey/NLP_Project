```python
import os

from xlutils.copy import copy
import xlwt
import xlrd
from xlrd import open_workbook

from openpyxl import Workbook
from openpyxl import load_workbook
import openpyxl

import win32com.client as win32
```

```python
def cal_mean_var(path):
    rb = open_workbook (path)
    r_sheet = rb.sheet_by_index(0)
    wb = copy(rb)
    w_sheet = wb.get_sheet(0)
    row = r_sheet.nrows-2
    
    w_sheet.write(row, 0, "Mean")
    w_sheet.write(row, 1, xlwt.Formula('AVERAGE(B2:B' + str(row) + ')'))
    w_sheet.write(row, 2, xlwt.Formula('AVERAGE(C2:C' + str(row) + ')'))
    w_sheet.write(row, 3, xlwt.Formula('AVERAGE(D2:D' + str(row) + ')'))
    w_sheet.write(row, 4, xlwt.Formula('AVERAGE(E2:E' + str(row) + ')'))
    w_sheet.write(row, 5, xlwt.Formula('AVERAGE(F2:F' + str(row) + ')'))
    w_sheet.write(row, 6, xlwt.Formula('AVERAGE(G2:G' + str(row) + ')'))
    w_sheet.write(row, 7, xlwt.Formula('AVERAGE(H2:H' + str(row) + ')'))
    w_sheet.write(row, 8, xlwt.Formula('AVERAGE(I2:I' + str(row) + ')'))
    w_sheet.write(row, 9, xlwt.Formula('AVERAGE(J2:J' + str(row) + ')'))
    w_sheet.write(row, 10, xlwt.Formula('AVERAGE(K2:K' + str(row) + ')'))
    w_sheet.write(row, 11, xlwt.Formula('AVERAGE(L2:L' + str(row) + ')'))
    w_sheet.write(row, 12, xlwt.Formula('AVERAGE(M2:M' + str(row) + ')'))

    w_sheet.write(row + 1, 0, "Var")
    w_sheet.write(row + 1, 1, xlwt.Formula('VAR(B2:B' + str(row) + ')'))
    w_sheet.write(row + 1, 2, xlwt.Formula('VAR(C2:C' + str(row) + ')'))
    w_sheet.write(row + 1, 3, xlwt.Formula('VAR(D2:D' + str(row) + ')'))
    w_sheet.write(row + 1, 4, xlwt.Formula('VAR(E2:E' + str(row) + ')'))
    w_sheet.write(row + 1, 5, xlwt.Formula('VAR(F2:F' + str(row) + ')'))
    w_sheet.write(row + 1, 6, xlwt.Formula('VAR(G2:G' + str(row) + ')'))
    w_sheet.write(row + 1, 7, xlwt.Formula('VAR(H2:H' + str(row) + ')'))
    w_sheet.write(row + 1, 8, xlwt.Formula('VAR(I2:I' + str(row) + ')'))
    w_sheet.write(row + 1, 9, xlwt.Formula('VAR(J2:J' + str(row) + ')'))
    w_sheet.write(row + 1, 10, xlwt.Formula('VAR(K2:K' + str(row) + ')'))
    w_sheet.write(row + 1, 11, xlwt.Formula('VAR(L2:L' + str(row) + ')'))
    w_sheet.write(row + 1, 12, xlwt.Formula('VAR(M2:M' + str(row) + ')'))
    
    w_sheet.write(row, 13, xlwt.Formula('AVERAGE(N2:N' + str(row) + ')'))
    w_sheet.write(row, 14, xlwt.Formula('AVERAGE(O2:O' + str(row) + ')'))
    w_sheet.write(row, 15, xlwt.Formula('AVERAGE(P2:P' + str(row) + ')'))
    w_sheet.write(row, 16, xlwt.Formula('AVERAGE(Q2:Q' + str(row) + ')'))
    
    w_sheet.write(row + 1, 13, xlwt.Formula('VAR(N2:N' + str(row) + ')'))
    w_sheet.write(row + 1, 14, xlwt.Formula('VAR(O2:O' + str(row) + ')'))
    w_sheet.write(row + 1, 15, xlwt.Formula('VAR(P2:P' + str(row) + ')'))
    w_sheet.write(row + 1, 16, xlwt.Formula('VAR(Q2:Q' + str(row) + ')'))
    
    col = r_sheet.ncols
    
    w_sheet.write(0, col, "Risk/Exposure")
    for i in range(1, row):
        if r_sheet.cell_value(i, 4) != 0:
            w_sheet.write(i, col, xlwt.Formula('G' + str(i+1) + '/E' + str(i+1)))
            
    w_sheet.write(0, col+1, "PSentiment/Exposure")
    for i in range(1, row):
        if r_sheet.cell_value(i, 4) != 0:
            w_sheet.write(i, col+1, xlwt.Formula('I' + str(i+1) + '/E' + str(i+1)))
                      
    w_sheet.write(0, col+2, "NSentiment/Exposure")
    for i in range(1, row):
        if r_sheet.cell_value(i, 4) != 0:
            w_sheet.write(i, col+2, xlwt.Formula('K' + str(i+1) + '/E' + str(i+1)))
                      
    w_sheet.write(0, col+3, "Sentiment/Exposure")
    for i in range(1, row):
        if r_sheet.cell_value(i, 4) != 0:
            w_sheet.write(i, col+3, xlwt.Formula('M' + str(i+1) + '/E' + str(i+1)))

    wb.save(path)
```

```python
def convert_to_xls (path):
    fname = path #full to .xls file
    #fname = "E:\\NLP_Dataset\\test" + "\\" + i + "\\" + i + "_Results.xls"
    excel = win32.gencache.EnsureDispatch('Excel.Application')
    wb = excel.Workbooks.Open(fname)

    wb.SaveAs(fname.replace("xlsx", "xls"), FileFormat = 56)    #FileFormat = 51 is for .xlsx extension
    wb.Close()                               #FileFormat = 56 is for .xls extension
    excel.Application.Quit()
```

```python
def get_mean_summary(source, path):
    wb = openpyxl.load_workbook (path)
    w_sheets = wb.get_sheet_names()
    w_sheet = wb.get_sheet_by_name(w_sheets[0])
    class_list = list(os.listdir(source))
    write_row = 1
    for i in class_list:
        sr = openpyxl.load_workbook (source + '\\' + i + '\\' + i + '_Results.xlsx', data_only = True)
        s_sheets = sr.get_sheet_names()
        s_sheet = sr.get_sheet_by_name(s_sheets[0])
        row = s_sheet.max_row
        w_sheet.cell(write_row, 1).value = i
        for j in range(2,18):
            s_cell = s_sheet.cell(row-1, j).value
            #print(s_cell)
            w_sheet.cell(write_row, j).value = s_cell
        write_row = write_row + 1
    wb.save(path)
    sr.close()
    wb.close()
```

```python
def get_var_summary(source, path):
    wb = openpyxl.load_workbook (path)
    w_sheets = wb.get_sheet_names()
    w_sheet = wb.get_sheet_by_name(w_sheets[0])
    class_list = list(os.listdir(source))
    write_row = 1
    for i in class_list:
        sr = openpyxl.load_workbook (source + '\\' + i + '\\' + i + '_Results.xlsx', data_only = True)
        s_sheets = sr.get_sheet_names()
        s_sheet = sr.get_sheet_by_name(s_sheets[0])
        row = s_sheet.max_row
        w_sheet.cell(write_row, 1).value = i
        for j in range(2,18):
            s_cell = s_sheet.cell(row, j).value
            #print(s_cell)
            w_sheet.cell(write_row, j).value = s_cell
        write_row = write_row + 1
    wb.save(path)
    sr.close()
    wb.close()
```
